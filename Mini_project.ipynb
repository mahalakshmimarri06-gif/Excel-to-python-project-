### Category-wise Sales vs. Profitability Analysis

```python
category_analysis = df.groupby('Category').agg(
    TotalSales=('TotalAmount', 'sum'),
    TotalNetSalesRevenue=('NetSalesRevenue', 'sum')
).reset_index()

category_analysis['ProfitabilityRatio'] = category_analysis['TotalNetSalesRevenue'] / category_analysis['TotalSales']

print("Category Analysis (Sorted by Total Sales):")
display(category_analysis.sort_values(by='TotalSales', ascending=False))

print("\nCategory Analysis (Sorted by Profitability Ratio):")
display(category_analysis.sort_values(by='ProfitabilityRatio', ascending=True))
```

### Product-wise Sales vs. Profitability Analysis

```python
product_analysis = df.groupby('ProductName').agg(
    TotalSales=('TotalAmount', 'sum'),
    TotalNetSalesRevenue=('NetSalesRevenue', 'sum')
).reset_index()

product_analysis['ProfitabilityRatio'] = product_analysis['TotalNetSalesRevenue'] / product_analysis['TotalSales']

print("Product Analysis (Top 10 by Total Sales):")
display(product_analysis.sort_values(by='TotalSales', ascending=False).head(10))

print("\nProduct Analysis (Bottom 10 by Profitability Ratio):")
display(product_analysis.sort_values(by='ProfitabilityRatio', ascending=True).head(10))
```

### Possible Reasons for High Sales, Low Profitability

If a category or product shows high sales volume but a low profitability ratio, several factors could be at play:

1.  **High Discounts:** Aggressive discounting might boost sales but severely cut into profit margins. This is common for products in highly competitive markets or during promotional events.
2.  **High Shipping Costs:** If a product is bulky, heavy, or requires special handling, its shipping costs might be disproportionately high compared to its selling price, reducing profitability.
3.  **High Return Rates:** Products with frequent returns or exchanges can incur additional shipping, restocking, and processing costs, thereby eroding profits despite initial sales.
4.  **Low Unit Price with High Associated Costs:** Some products might have a low unit price but still incur significant fixed costs (e.g., marketing, customer service, handling) per unit, making them less profitable than higher-priced items.
5.  **Competitive Pricing Pressure:** In a saturated market, businesses might be forced to lower prices to match competitors, leading to thin profit margins even with high sales.
6.  **Operational Inefficiencies:** Inefficient supply chain management, high storage costs, or poor inventory management can also eat into profits.
7.  **Tax Structure:** Some products might be subject to higher taxes, which, if not fully passed on to the customer, can reduce net revenue.
